```python
import pydotplus
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import Image
from sklearn import tree
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python

from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Data/guess_who.csv")
encoder = LabelEncoder()

target = df['name']

new_df = df.drop(columns='name')

features = pd.DataFrame()

for col in new_df:
  encoded_data = encoder.fit_transform(new_df[col])
  features[col] = encoded_data.tolist()


print(features)
```

Create dicision tree classifier object

```python
decisiontree = DecisionTreeClassifier(random_state=0)
```

Train model

```python
model = decisiontree.fit(features, target)
```

Create DOT data

```python
feature_names = features.columns
print(feature_names)
print(target)
```

```python
dot_data = tree.export_graphviz(decisiontree,
                               out_file=None,
                               feature_names=feature_names,
                               class_names=[target.name])
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```
